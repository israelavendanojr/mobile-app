```python
import cv2
import mediapipe as mp
import numpy as np

# Setup MediaPipe, drawing for visualization, pose grabs pose estimation model
mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose
```

```python
# VIDEO CAPTURE

# Grabbing video capture device, 1 is the second camera (webcam)
cap = cv2.VideoCapture(1)

# Loop through feed
while cap.isOpened():
    # Read feed, ret is return value, frame is the image
    ret, frame = cap.read()
    # Gives us a window to display the feed
    cv2.imshow('MediaPipe Feed', frame)
    
    # Press 'q' to quit, 10 is the delay (ms), 0xFF is last key pressed
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

# Release the capture
cap.release()
cv2.destroyAllWindows()
```

```python
# MAKING DETECTIONS

cap = cv2.VideoCapture(1)
with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
    while cap.isOpened():
        ret, frame = cap.read()
        
        # Recolor image, BGR to RGB
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False
        
        # Flip image
        image = cv2.flip(image, 1)

        # Make detection
        results = pose.process(image)

        # Recolor image back, RGB to BGR
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
        # Render detections and show feed
        mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)
        cv2.imshow('MediaPipe Feed', image)
        
        # Press 'q' to quit, 10 is the delay (ms), 0xFF is last key pressed
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

# Release the capture
cap.release()
cv2.destroyAllWindows()
```

```python
# CALCULATE JOINT ANGLES

def calculate_angle(a, b, c):
    a = np.array(a) # First
    b = np.array(b) # Mid
    c = np.array(c) # End
    
    # Calculate angle
    radians = np.arctan2(c[1] - b[1], c[0] - b[0]) - np.arctan2(a[1] - b[1], a[0] - b[0])
    angle = np.abs(radians * 180.0 / np.pi)
    
    # Check if angle is greater than 180,
    if angle > 180.0:
        angle = 360 - angle
    
    return angle
```

```python
# preprocessing landmarks, extracting only x and y coordinates
right_shoulder = [landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x, landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y]
right_elbow = [landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].x, landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].y]
right_wrist = [landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].x, landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].y]

# calculate angle
calculate_angle(right_shoulder, right_elbow, right_wrist)
```

```python
# REAL TIME ANGLE DETECTION

cap = cv2.VideoCapture(1)
with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
    while cap.isOpened():
        ret, frame = cap.read()
        
        # Recolor image, BGR to RGB
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False
        
        # Get image height and width for drawing
        image_height, image_width = image.shape[:2]

        # Make detection
        results = pose.process(image)

        # Recolor image back, RGB to BGR
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

        # Extract landmarks
        try:
            landmarks = results.pose_landmarks.landmark

            # Extract and preprocess landmarks for bicep curl
            right_shoulder = [landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x, landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y]
            right_elbow = [landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].x, landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].y]
            right_wrist = [landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].x, landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].y]
            
            # Calculate angle
            angle = calculate_angle(right_shoulder, right_elbow, right_wrist)
            
            # Visualize angle
            cv2.putText(image, str(angle), tuple(np.multiply(right_elbow, [image_width, image_height]).astype(int)), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA)
        except:
            pass
        
        # Render detections and show feed
        mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)
        cv2.imshow('MediaPipe Feed', image)
        
        # Exit on 'q'
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

# Release the capture
cap.release()
cv2.destroyAllWindows()
```

```python
# BICEP CURL COUNTER

rep_count = 0
rep_stage = None

cap = cv2.VideoCapture(1)
with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
    while cap.isOpened():
        ret, frame = cap.read()
        
        # Recolor image, BGR to RGB
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False
        
        # Get image height and width for drawing
        image_height, image_width = image.shape[:2]

        # Make detection
        results = pose.process(image)

        # Recolor image back, RGB to BGR
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

        # Extract landmarks
        try:
            landmarks = results.pose_landmarks.landmark

            # Extract and preprocess landmarks for bicep curl
            right_shoulder = [landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x, landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y]
            right_elbow = [landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].x, landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].y]
            right_wrist = [landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].x, landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].y]
            
            # Calculate angle
            angle = calculate_angle(right_shoulder, right_elbow, right_wrist)
            
            # Visualize angle
            cv2.putText(image, str(angle), tuple(np.multiply(right_elbow, [image_width, image_height]).astype(int)), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 3, cv2.LINE_AA)

            # Counter logic
            if angle > 145:
                stage = "concentric"
            if angle < 60 and stage == "concentric":
                stage = "eccentric"
                rep_count += 1
        except:
            pass
        
        # BICEP CURL COUNTER RENDERING

        # Define box position and size
        box_x, box_y = 20, 20
        box_width, box_height = 180, 90

        # Draw background rectangle
        cv2.rectangle(image, (box_x, box_y), (box_x + box_width, box_y + box_height), (0, 0, 0), -1)       # filled black box
        cv2.rectangle(image, (box_x, box_y), (box_x + box_width, box_y + box_height), (0, 255, 0), 2)      # green border

        # Draw REPS label
        cv2.putText(
            image,
            'REPS',
            (box_x + 10, box_y + 30),
            cv2.FONT_HERSHEY_SIMPLEX,
            0.7,
            (0, 255, 0),
            2,
            cv2.LINE_AA
        )

        # Draw rep count
        cv2.putText(
            image,
            str(rep_count),
            (box_x + 10, box_y + 65),
            cv2.FONT_HERSHEY_SIMPLEX,
            1.5,
            (255, 255, 255),
            2,
            cv2.LINE_AA
        )

        # Draw stage label
        if 'stage' in locals() and stage is not None:
            cv2.putText(
                image,
                stage.upper(),
                (box_x + 100, box_y + 65),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.8,
                (0, 255, 255),  # yellow stage
                2,
                cv2.LINE_AA
            )

        # Render detections and show feed
        mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)
        cv2.imshow('MediaPipe Feed', image)

        # Exit on 'q'
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

# Release the capture
cap.release()
cv2.destroyAllWindows()
```
